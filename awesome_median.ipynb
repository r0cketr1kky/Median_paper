```
import numpy as np
import random
import matplotlib.pyplot as plt
```

```
class User:
  def __init__(self, id, inp):
    self.partner_id = id
    self.input1 = inp
    self.ran = np.random.randint(-2**16, 2**16 - 1)
```

```
class Computations:

  def __init__(self, user: User):
    self.user = user
  
  def operation1(self):
    out_inp1 = self.user.input1 + self.user.ran
    return out_inp1
      
  def operation2(self):
    out_inp2 = self.user.input1 - self.user.ran
    return out_inp2  
```

```
l = []
```

```
def pairwiseBlind(seq):
	n = np.zeros(len(seq))
	for i in range(0, len(seq), 2):
		r = np.random.randint(-2**16, 2**16)
		n[i] = (seq[i] + r)
		n[i+1] = (seq[i+1] - r)
	return n
```

```
def find_median(users_list, l, rang):
  
  for i in range(rang):
    x = users_list.pop(random.randint(0,len(users_list)-1)) # RANDOMLY POP AN ELEMENT FROM USERS_LIST
    y = users_list.pop(random.randint(0,len(users_list)-1)) # RANDOMLY POP AN ELEMENT FROM USERS_LIST

  #print(x.input1, y.input1)
  #print(len(users_list))

    y.ran = x.ran

    val1 = Computations(x).operation1()
    val2 = Computations(y).operation2()

    l.append(val1) # APPENDING VAL1 TO LIST L1
    l.append(val2) # APPENDING VAL2 TO LIST L2

  return
```

```
def streamingMedian(seq):

	random.shuffle(seq)
 
	m = 0

	for nextElt in seq:
		if m > nextElt:
			m -= 1
		elif m < nextElt:
			m += 1

	return m
```

```
beta_med = []
```

```
def beta_run(): 
  
  for i in range(1, 25):
    beta_values = np.random.beta(2**16, 2**16, 2**i)
    
    print(i)
    #print('vector size: 2^%d' % i)
    #print('true median: %s' % np.median(beta_values))
    #print('approx median: %s' % streamingMedian(beta_values))

    l = pairwiseBlind(beta_values)
    m = streamingMedian(l)

    #print('blinded approx median: %s' % m)
    #print('blinded true median %s' % np.median(l))
    #print('---------------------------')
  

    beta_med.append(m)

  #true_median = np.median(beta_values)
  #print(true_median)

  #plt.plot(lmed)
  #plt.show()

  #print(l)

  return
```

```
beta_run()
```

```
for _ in range(5):
  beta_run()
```
